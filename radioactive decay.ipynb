```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(N_t,t):
    dN_tdt =- k*N_t
    return dN_tdt

N0=2000                        #where N_t is the number of undecayed atoms present at any instant of time
k=0.5                           # No
t=np.linspace(0,50)              # k is called decay constant
N_t=odeint(model,N0,t)

plt.plot(t,N_t)
plt.xlabel('time')
plt.ylabel('N(t)')
plt.show()


```
