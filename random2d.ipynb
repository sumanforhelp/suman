Random walk in 2D corresponds to a random process in which there is value at a point on 2D and its value changes randomly with time. If it moves in x-direction then y value remains same and vice-versa. Similarly if point moves obliquely in the plane then both x and y coodinate change. It has wide range of importance like a tiger moving in jungle for food, stock-exchange,diffusion etc and most probably spread of corona virus in a community can also be modeled by random walk.


Ok! How do we do it?
1.First, choose the starting point to be {x=0,y=0}.
2.Generate the random number. use if...........else statement to determine the next point in the 2D plane. 
3. Here we will consider the movement only along +ve X or +ve Y or -ve X or -ve Y.
4. In which direction to move is determined by generated random number and if......else statement


```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
import random as random
import numpy as np
import pandas as pd
sns.set()
isprint=True
```

```python
walk2D = open("data/randomwalk2d.txt",'w') 
```

```python
x=0
y=0
X=[x]
Y=[y]
```

```python
for u in range(1,100000):
    r=random.uniform(1,5)
    if r>1 and r<2:
        x=X[u-1]+1
        y=Y[u-1]
        X.append(x);Y.append(y)
    
    elif r>2 and r<3:
        x=X[u-1]-1
        y=Y[u-1]
        X.append(x);Y.append(y)
    elif r>3 and r<4:
        y=Y[u-1]+1
        x=X[u-1]
        X.append(x);Y.append(y)
    
    elif r>4 and r<5:
        y=Y[u-1]-1
        x=X[u-1]
        X.append(x);Y.append(y)
        
    if isprint:
        #print("step: ",u,"random number is: ", r,\
                                #"coordinate is : ", x,y)

        walk2D.write("step: " + str(u) + " random number is: " + str(r) + 
                                 " coordinate is : " + str(x) + ","+str(y))
        walk2D.write("\n")
        
walk2D.close()
```
